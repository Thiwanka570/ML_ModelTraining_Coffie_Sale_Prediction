```python
import pandas as pd



# Load the data from 'coffee_sale.csv' into a pandas DataFrame.
try:
    df = pd.read_csv('/content/MyDataSet/Salary_dataset.csv')

    # Display the first 5 rows of the DataFrame to verify it loaded correctly.
    display(df.head())

    # Print the DataFrame's information to see data types and non-null counts.
    display(df.info())

except FileNotFoundError:
    print("Could not find 'coffee_sale.csv'. Please ensure the file is in the correct directory.")
    df = None # Set df to None to indicate loading failed
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
df.dropna(subset=['Unnamed'], inplace=True)
```

# Task
Train a Linear Regression model on the dataset to predict 'Salary' based on 'YearsExperience', and then use the trained model to predict salaries.

## Data cleaning

### Subtask:
Remove the 'Unnamed: 0' column as it appears to be an index and not a feature.


**Reasoning**:
Remove the 'Unnamed: 0' column from the DataFrame as it is an index column and display the head of the dataframe to verify.



```python
df.drop('Unnamed: 0', axis=1, inplace=True)
display(df.head())
```

## Model selection

### Subtask:
Choose a suitable model for regression, such as Linear Regression.


## Model training

### Subtask:
Train the selected Linear Regression model using 'YearsExperience' as the feature and 'Salary' as the target variable.


**Reasoning**:
Import LinearRegression, instantiate the model, define X and y, and fit the model to the data as per the instructions.



```python
from sklearn.linear_model import LinearRegression

# Instantiate a LinearRegression object.
model = LinearRegression()

# Define the feature variable X and the target variable y.
X = df[['YearsExperience']]  # X must be a 2D array
y = df['Salary']

# Fit the Linear Regression model to the data.
model.fit(X, y)

print("Model training complete.")
```

## Prediction

### Subtask:
Use the trained model to predict salaries based on new 'YearsExperience' values.


**Reasoning**:
Define new 'YearsExperience' values and use the trained model to predict salaries for these values.



```python
import numpy as np

# Define new 'YearsExperience' values.
new_years_experience = np.array([1.5, 5.5, 11.0]).reshape(-1, 1) # Reshape for the model's input

# Use the trained model to predict salaries for the new values.
predicted_salaries = model.predict(new_years_experience)

# Display the predicted salaries.
print("Predicted salaries for new YearsExperience values:")
for years, salary in zip(new_years_experience.flatten(), predicted_salaries):
    print(f"Years of Experience: {years}, Predicted Salary: {salary:.2f}")
```

## Summary:

*   The 'Unnamed: 0' column, likely an index, was removed from the dataset.
*   Linear Regression was selected as the appropriate model due to the observed linear relationship between 'YearsExperience' and 'Salary'.
*   A Linear Regression model was successfully trained using 'YearsExperience' as the feature and 'Salary' as the target variable.
*   The trained model predicted salaries for new 'YearsExperience' values: \$39023.15 for 1.5 years, \$76823.00 for 5.5 years, and \$128797.79 for 11.0 years.
